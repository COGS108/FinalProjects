# An exploratory analysis on optimal desalination plants in California

This notebook will be exploring the hypothetical analysis of potential placement locations for desalination plants along the coast of California. 

# Introduction

## Background

Seawater desalination *“is a process that turns ocean water into fresh drinking water by removing salt and impurities. The process involves either reverse osmosis membrane or thermal distillation (evaporation) technologies”* (1).
While the lack of Water has always been a concern for the world, there are more than 21,000 desalination facilities operating worldwide and 300 million people are using the freshwater from more than 17,000 desalination plants in 150 countries (2). Although California has been under the situation of drought for more than five years, only 14 percent of the Unite States’ desalination plants are placed in California, according to the Texas Water Development Board (3). Although our ideal will be supplying all the california population's water consumption, the cost of building a desalination plant is very high. For example, it costs 32 million to build a 2.5 MGD seawater desalination plant, and approximately $658 million is needed to build a 100 MGD seawater desalination plant in Texas (3). Therefore, in our project, we are interested in find a way to provide the desalinated water to the maximum population effectively by adjusting the location we place the desalination plants.

### What is Desalination?

## Hypothesis 

Given that the size of California is around 160,000 miles and that any given desalination plant produces water on the order of 10^7 - 10^8 gallons per day. This is an estimation produced by Fermi Estimation to predict water size. We plan on exploring a finite number of desalination plants.


### Fermi Estimation 
** What is the approximate size of a desalination plant? **

If a house is about 2,000 square feet, and a factory plant is probably 10x the size of a house, making it 20,000 square feet and approximately 4 miles. 

** How many counties in California near the coast? ** 

We then only take in account the counties with large enough water masses to place desalination plants. There are close to 50 counties in California, not of all have large enough bodies of water. We can assume there are at least 40 percent of California has some body of water large enough to place a plant. In other words, there should be at least 20 counties close to bodies of water. 

** How big are the counties along the California coast? ** 

Each county is a different size, but if there are 160,000 miles California, and we're stating 40 percent of California has water, then there is 64,000 miles of area that potentially have coastal areas. 

** What is the size of area in each coastal county that have water? **

We can assume that of the 64,000 miles of area, each county has less than 30 percent coastal space. In result, there is about 19,000 miles of area that are available for desalination plants. 

** What is the maximum desalination plants possible placed on the coast of California ** 

If we ever decided to place a plant every single place along the coast we would be able to get approximately:

> 
19000 miles / 4 miles = 4750 plants along the coast.

````python
def main():

    # How many points are in our dataset?
    num_points = 743

    # For each of those points how many dimensions do they have?
    # Note: Plotting will only work in two or three dimensions
    dimensions = 2

    # Bounds for the values of those points in each dimension
    lower = -125
    upper = 42

    # The K in k-means. How many clusters do we assume exist?
    num_clusters = 54

    # When do we say the optimization has 'converged' and stop updating clusters
    cutoff = 0.2

    # Generate some points to cluster
    coordlist = pd.DataFrame([newdf['INTPTLAT'],newdf['INTPTLONG']])
    coordlist = coordlist.transpose()
    coordlist = coordlist.values.tolist()

    # Generate some points to cluster
    points = [
        makePoint(coords[0],coords[1]) for coords in coordlist
    ]

    # Cluster those data!
    clusters = kmeans(points, num_clusters, cutoff)
    output = pd.DataFrame()
    # Print our clusters
    for i, c in enumerate(clusters):
        for p in c.points:
            #print (" Cluster: ", i, "\t Point :", p, "Pop of cluster :", findPopofCluster(c))
            pointdf = pd.DataFrame(data=[i,p.coords[0],p.coords[1],c.calculateCentroid().coords])
            output = output.append(pointdf.transpose())
    output.to_csv("kmeansresult.csv")
    # Display clusters using plotly for 2d data
class Point(object):
    '''
    A point in n dimensional space
    '''
    def __init__(self, coords):
        '''
        coords - A list of values, one per dimension
        '''

        self.coords = coords
        self.n = len(coords)

    def __repr__(self):
        return str(self.coords)

class Cluster(object):
    '''
    A set of points and their centroid
    '''

    def __init__(self, points):
        '''
        points - A list of point objects
        '''

        if len(points) == 0:
            raise Exception("ERROR: empty cluster")

        # The points that belong to this cluster
        self.points = points

        # The dimensionality of the points in this cluster
        self.n = points[0].n

        # Assert that all points are of the same dimensionality
        for p in points:
            if p.n != self.n:
                raise Exception("ERROR: inconsistent dimensions")

        # Set up the initial centroid (this is usually based off one point)
        self.centroid = self.calculateCentroid()

    def __repr__(self):
        '''
        String representation of this object
        '''
        return str(self.points)

    def update(self, points):
        '''
        Returns the distance between the previous centroid and the new after
        recalculating and storing the new centroid.
        Note: Initially we expect centroids to shift around a lot and then
        gradually settle down.
        '''
        old_centroid = self.centroid
        self.points = points
        self.centroid = self.calculateCentroid()
        shift = getDistance(old_centroid, self.centroid)
        return shift

    def calculateCentroid(self):
        '''
        Finds a virtual center point for a group of n-dimensional points
        '''
        numPoints = len(self.points)
        # Get a list of all coordinates in this cluster
        coords = [p.coords for p in self.points]
        # Reformat that so all x's are together, all y'z etc.
        unzipped = zip(*coords)
        # Calculate the mean for each dimension
        centroid_coords = [math.fsum(dList)/numPoints for dList in unzipped]

        return Point(centroid_coords)

def kmeans(points, k, cutoff):

    # Pick out k random points to use as our initial centroids
    initial = random.sample(points, k)

    # Create k clusters using those centroids
    # Note: Cluster takes lists, so we wrap each point in a list here.
    clusters = [Cluster([p]) for p in initial]

    # Loop through the dataset until the clusters stabilize
    loopCounter = 0
    while True:
        # Create a list of lists to hold the points in each cluster
        lists = [[] for _ in clusters]
        clusterCount = len(clusters)

        # Start counting loops
        loopCounter += 1
        # For every point in the dataset ...
        for p in points:
            # Get the distance between that point and the centroid of the first
            # cluster.
            smallest_distance = getDistance(p, clusters[0].centroid)

            # Set the cluster this point belongs to
            clusterIndex = 0

            # For the remainder of the clusters ...
            for i in range(clusterCount - 1):
                # calculate the distance of that point to each other cluster's
                # centroid.
                distance = getDistance(p, clusters[i+1].centroid)
                # If it's closer to that cluster's centroid update what we
                # think the smallest distance is
                if distance < smallest_distance:
                    smallest_distance = distance
                    clusterIndex = i+1
            # After finding the cluster the smallest distance away
            # set the point to belong to that cluster
            lists[clusterIndex].append(p)

        # Set our biggest_shift to zero for this iteration
        biggest_shift = 0.0

        # For each cluster ...
        for i in range(clusterCount):
            # Calculate how far the centroid moved in this iteration
            shift = clusters[i].update(lists[i])
            # Keep track of the largest move from all cluster centroid updates
            biggest_shift = max(biggest_shift, shift)

        # If the centroids have stopped moving much, say we're done!
        if biggest_shift < cutoff:
            print ("Converged after %s iterations" % loopCounter)
            break
    return clusters

def getDistance(a, b):
    return haversine(a.coords,b.coords)

def makePoint(lat,long):
    '''
    Returns a Point object with n dimensions and values between lower and
    upper in each of those dimensions
    '''
    p = Point([lat,long])
    return p
def findPopofPoint(p):
    index = newdf[(newdf['INTPTLAT'] == p.coords[0]) & (newdf['INTPTLONG'] == p.coords[1])].index.tolist()
    return newdf.iloc[index[0]]['2010POP']
def findPopofCluster(c):
    popCluster = 0 
    it = 0
    for i in c.points:
        popCluster += findPopofPoint(c.points[it])
        it += 1
    return popCluster
if __name__ == "__main__":
    main()
    
````

### Data Stats

By using the [tutorial](http://www.christianpeccei.com/zipmap/) by Christian Peccei,
we want to see how much of California have zip codes that include coastal areas:

According to Christian, the following code 
> reads polygon data from an ASCII boundary file. Returns a dictionary with polygon IDs for keys. The value for each key is another dictionary with three keys:
- *'name' - the name of the polygon*
- *'polygon' - list of (longitude, latitude) pairs defining the main polygon boundary *
- *'exclusions' - list of lists of (lon, lat) pairs for any exclusions in the main polygon*"

````python
def read_ascii_boundary(filestem):
    metadata_file = filestem + 'a.dat'
    data_file = filestem + '.dat'
    # Read metadata
    lines = [line.strip().strip('"') for line in open(metadata_file)]
    polygon_ids = lines[::6]
    polygon_names = lines[2::6]
    polygon_data = {}
    for polygon_id, polygon_name in zip(polygon_ids, polygon_names):
        # Initialize entry with name of polygon.
        # In this case the polygon_name will be the 5-digit ZIP code.
        polygon_data[polygon_id] = {'name': polygon_name}
    del polygon_data['0']
    # Read lon and lat.
    f = open(data_file)
    for line in f:
        fields = line.split()
        if len(fields) == 3:
            # Initialize new polygon
            polygon_id = fields[0]
            polygon_data[polygon_id]['polygon'] = []
            polygon_data[polygon_id]['exclusions'] = []
        elif len(fields) == 1:
            # -99999 denotes the start of a new sub-polygon
            if fields[0] == '-99999':
                polygon_data[polygon_id]['exclusions'].append([])
        else:
            # Add lon/lat pair to main polygon or exclusion
            lon = float(fields[0])
            lat = float(fields[1])
            if polygon_data[polygon_id]['exclusions']:
                polygon_data[polygon_id]['exclusions'][-1].append((lon, lat))
            else:
                polygon_data[polygon_id]['polygon'].append((lon, lat))
    return polygon_data
````

### Probable Confounds

- If we are placing the desalination plant in a more population dense area, the water delivery will be more efficient, but the desalination plant itself will take up a huge space. Therefore placing the plant in the most population dense area might not be the best idea, and we need to come up with a constraint of population density for where we can place the plants.
- Since the implementation of desalination plants takes relatively long time, and our project is using population as a independent variable, we might also need to consider the growth and decay of population, or the population shifts in California 
- Water pipe distance and cost to maintain to transport water to plants would also cause the number of plants to change 
- Since we are finding the centroid of each cluster and moving the centroid to the coastal area, the moved centroid might not be the point that minimize the water delivery distance any more. Also, we still need to be checked if the moved centroid points are the realistic places to place the desalination plants. For example, environmental factors such as cliff sides, muddy areas or protected animal habitats. 
- We also might have inaccurate results since we got our California beach data from Wikipedia, versus a database.  
 

### Other thoughts on alternative methods
If our methods don’t work, such as if we get a strangely small amount of desalination plants, we might have data that is better for answering a similar related question or our datasets might not be the most accurate. Find more suitable datasets is one way we could try to “debug” this problem. Also in the case that our method doesn’t give an accurate enough description, we might have to find new datasets that include more features to give us more potential undiscovered variables that we may not have considered prior to exploring this problem.
