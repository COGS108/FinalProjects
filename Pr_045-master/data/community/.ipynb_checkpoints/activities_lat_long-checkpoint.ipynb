```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
```

```python
df_activities = pd.read_csv('Activities.csv')
df_activities['Activity_Location'][69] = "8285 Skyline Dr\nSan Diego, CA, US 92114"
df_activities['Activity_Location'][69]
#df_activities[['Activity_Location'] == "8285 Skyline Dr\n-117.026223\nSan Diego, CA, US 92114"]
```

```python
df_activities = df_activities.dropna()
df_activities = df_activities.reset_index(drop=True)
df_activities
```

```python
def clean_address(addr):
    addr = addr.replace("\n", " ", 1)
    addr = addr[:-14]
    return addr

clean_address("8285 Skyline Dr\nSan Diego, CA, US 92114")
```

```python
geolocator = Nominatim()

clean_addresses = []
latitude = []
longitude = []

for address in df_activities['Activity_Location']:
    clean = clean_address(address)
    geo = geolocator.geocode(str(clean))
    if geo is None:
        lat = np.nan
        long = np.nan
    else:
        lat = geo.latitude
        long = geo.longitude
    
    clean_addresses.append(clean)
    latitude.append(lat)
    longitude.append(long)
df_activities['Clean_Addresses'] = clean_addresses
df_activities['Lattitude'] = latitude
df_activities['Longitude'] = longitude
```

```python
df_activities = df_activities.dropna()
df_activities = df_activities.reset_index(drop=True)
df_activities = df_activities.drop(['Activity_url','Activity_Location'], axis=1)
df_activities
```

```python
df_activities.to_csv('cleaned_activities.csv')
```
